# 実験を実施する

DESC

「初期セットアップを行う」を実行すると、DMPのデータセット構造(datasetStructure)に従い、実験に必要なフォルダ構成が実験パッケージに用意されます。<br>
データセット構造毎に用意されるフォルダ構成は以下です。

- with_codeを選択した場合
  - **図を用意(ci tests含む)**
- for_parametersを選択した場合
  - **図を用意**

各フォルダの使用について各データセット構造で共通する条件と特有の条件があります。<br>
下記の条件を満たすように、実験の準備をしてください。<br>

- 共通条件
  - ソースコードはPythonで記述し、`source`フォルダに配置すること  
  - 実験の入力データは`input_data`フォルダに配置すること  
  - 実験結果が`output_data`フォルダに出力されるように、実験ソースコードを記述すること  

- 特有条件
  - with_codeを選択した場合
    - なし

  - for_parametersを選択した場合
    - 各パラメータ実験の固有のパラメータファイルは、それぞれのparamsフォルダに配置すること
      パラメータファイルは任意の種類を使用可(.txt, .yaml など)
    - 実験結果は各パラメータ実験のoutput_dataフォルダに配置すること
    - **パラメータ実験を増やす方法**

実験結果が得られるまで、入力データや実験実行コードの修正を繰り返してください。

* 当環境の対象実験パッケージを確認する方法
* 当実験パッケージの範囲のデータしか同期されないこと
* パラメータ実験を削除したい場合の手順
* パラメータ実験名を変更したい場合　（初期設定やパラメータ実験作成後にもパラメータ実験名を更新できる旨を追記）
